```python
import matplotlib.pyplot as plt
import numpy as np
```

# Matplotlib default plot

```python
x=np.linspace(0, 2*np.pi)
y=np.sin(x)
z=-np.cos(x)

fig, ax=plt.subplots()

ax.plot(x, y, label=r'$\sin(x)$')

ax.set_xlabel(r"Value $x$")
ax.set_ylabel(r"Sine $\sin(x)$")
ax.set_title(r"An example plot of $\sin(x)$ without proper \TeX")
ax.legend()

plt.savefig("exampleSinglePlot_mpl_default.pdf")
```

# Nice Plot

```python
import niceplots.utils as pl


pl.initPlot()
```

```python
fig, ax=plt.subplots()
ax.plot(x, y, label=r'$\sin(x)$')
ax.set_xlabel(r"Value $x$")
ax.set_ylabel(r"Sine $\sin(x)$")

pl.finalizePlot(ax, title=r"An example plot of $\sin(x)$  with proper \TeX", outputFn="exampleSinglePlot_niceplots.pdf")
```
