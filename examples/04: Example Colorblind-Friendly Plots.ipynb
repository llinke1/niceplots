# Examples for color-blind friendly plots

Many viewers of a figure or readers of a paper have trouble distinguishing some colors. The default color scheme of `matplotlib` is not very accomodating for this, since it uses a red and green that look similar for people with Protanopia or Deuteranopia. To create accessible figures, `niceplots` automatically sets the default color scheme to a more color-blind friendly palette when running`nicepl.initPlot`. We also include options to quickly switch to other color-blind friendly palettes, as well as sequential and diverging colormaps. We show how to use these options in this notebook. 

```python
# Simple imports
import matplotlib.pyplot as plt
import numpy as np
import niceplots.utils as nicepl
```

```python
# Setting niceplot default settings
nicepl.initPlot()
```

```python
# Creating test data
N=8
x=np.linspace(0, np.pi)
data=[]
for i in range(N):
    data.append(np.sin(x)+i*0.1)
```

## Categorical Data

The preferred color scheme of a plot depends on the type of data that is shown. If the different lines shown, cannot be "ordered" in a sequential or diverging way, a categorical colorscheme is the best, as it does not imply a hierarchical relationship between the data series. 

### Default setting (`categorical1`, Petroff)

The default setting for `niceplots` is the categorical palette by Petroff with 8 colors (https://arxiv.org/pdf/2107.02270). It can be set with the option `colortype="categorical1"` in `initPlot` (or just leaving out `colortype`).

```python
fig, ax=plt.subplots()

ax.set_xlabel(r"Value $x$")
ax.set_ylabel(r"Function $f(x)$")

for i in range(N):
    offset=i*0.1
    ax.plot(x, data[i], label=r'$f(x)=\sin(x)+$'+f"${offset:.1f}$")

# Saving plot
nicepl.finalizePlot(ax, title=r"Example plot with default color palette (categorical1)", legendcols=2, outputFn="exampleColorblindfriendlyDefault.pdf")
```

### Option for categorical (`categorical2`, Okabe & Ito)

Another categorical palette is the one by Okabe & Ito (https://jfly.uni-koeln.de/color/#pallet). It can be set with the option `colortype="categorical2"` in `initPlot` or `setDefaultColors`.

```python
# Change default palette
nicepl.setDefaultColors(type="categorical2")


fig, ax=plt.subplots()

ax.set_xlabel(r"Value $x$")
ax.set_ylabel(r"Function $f(x)$")

for i in range(N):
    offset=i*0.1
    ax.plot(x, data[i], label=r'$f(x)=\sin(x)+$'+f"${offset:.1f}$")

# Saving plot
nicepl.finalizePlot(ax, title=r"Example plot with color palette `categorical2`", legendcols=2)
```

### Option for categorical (`categorical3`, Tol)

Another categorical palette is the one by Tol (https://cran.r-project.org/web/packages/khroma/vignettes/tol.html#introduction). It can be set with the option `colortype="categorical3"` in `initPlot` or `setDefaultColors`.

```python
# Change default palette
nicepl.setDefaultColors(type="categorical3")


fig, ax=plt.subplots()

ax.set_xlabel(r"Value $x$")
ax.set_ylabel(r"Function $f(x)$")

for i in range(N):
    offset=i*0.1
    ax.plot(x, data[i], label=r'$f(x)=\sin(x)+$'+f"${offset:.1f}$")

# Saving plot
nicepl.finalizePlot(ax, title=r"Example plot with color palette `categorical3`", legendcols=2)
```

## Sequential color palette

If the different data series shown have a clear relationship (i.e. increasing a common parameter) the best color choice is a sequential color palette. The colors can be directly drawn from a color palette, the default for `niceplots` is `copper`.


```python
# Change default palette
nicepl.setDefaultColors(type="sequential")


fig, ax=plt.subplots()

ax.set_xlabel(r"Value $x$")
ax.set_ylabel(r"Function $f(x)$")

for i in range(N):
    offset=i*0.1
    ax.plot(x, data[i], label=r'$f(x)=\sin(x)+$'+f"${offset:.1f}$")

# Saving plot
nicepl.finalizePlot(ax, title=r"Example plot with sequential color palette", legendcols=2)
```

For sequential color palettes, we can change the number of colors to draw from the color palette with the parameter `N` of `setDefaultColors`

```python
# Change default palette
nicepl.setDefaultColors(type="sequential", N=4)


fig, ax=plt.subplots()

ax.set_xlabel(r"Value $x$")
ax.set_ylabel(r"Function $f(x)$")

for i in range(N//2):
    offset=i*0.1
    ax.plot(x, data[i], label=r'$f(x)=\sin(x)+$'+f"${offset:.1f}$")

# Saving plot
nicepl.finalizePlot(ax, title=r"Example plot with sequential color palette", legendcols=2)
```

## Diverging color palette

If the different data series shown have a clear relationship and a threshold value (i.e. a common parameter is partly below zero, partly above zero) a diverging color palette can be appropriate. The colors can be directly drawn from a color palette, the default for `niceplots` is `coolwarm`.


```python
# Change default palette
nicepl.setDefaultColors(type="diverging")


fig, ax=plt.subplots()

ax.set_xlabel(r"Value $x$")
ax.set_ylabel(r"Function $f(x)$")

for i in range(N):
    offset=i*0.1
    ax.plot(x, data[i], label=r'$f(x)=\sin(x)+$'+f"${offset:.1f}$")

# Saving plot
nicepl.finalizePlot(ax, title=r"Example plot with diverging color palette", legendcols=2)
```
