```python
import matplotlib.pyplot as plt
import numpy as np
import niceplots.utils as nicepl
```

```python
nicepl.initPlot()
```

```python
x=np.linspace(0, 2*np.pi)

f1=np.sin(x)
y1=f1+0.3*(np.random.uniform(size=len(x))-0.5)

f2=np.cos(x)
y2=f2+0.3*(np.random.uniform(size=len(x))-0.5)
```

```python
colors=nicepl.getColorList(2, cmap='copper')
```

```python
gs_kw=dict(height_ratios=[3,1])

fig, axes=plt.subplots(nrows=2, ncols=2, gridspec_kw=gs_kw, sharex=True,
                       sharey="row", figsize=(7,4))
plt.subplots_adjust(hspace=0, wspace=0, 
                    left=0.13, right=0.99, bottom=0.13)

axes[1][0].set_xlabel("Value $x$")
axes[1][1].set_xlabel("Value $x$")


axes[0][0].set_ylabel("Function $f(x)$")
axes[0][0].plot(x, f1, label='$\sin(x)$', c=colors[0])
axes[0][0].scatter(x, y1, label='data $y$', c=colors[1])

axes[0][0].legend()

axes[0][1].plot(x, f2, label='$\cos(x)$', c=colors[0])
axes[0][1].scatter(x, y2, label='data $y$', c=colors[1])

axes[0][1].legend()

axes[1][0].set_ylabel("$y-f(x)$")
axes[1][0].axhline(0, c=colors[0], ls=':')
axes[1][0].scatter(x, y1-f1, c=colors[1])
axes[1][1].axhline(0, c=colors[0], ls=':')
axes[1][1].scatter(x, y2-f2, c=colors[1])


fig.suptitle(r"An example plot with multiple panels")

nicepl.finalizePlot(axes, tightlayout=False, showlegend=False,
                    outputFn="exampleMultiplePanels_niceplots.pdf")
```
